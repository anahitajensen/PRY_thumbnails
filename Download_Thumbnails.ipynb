```python
from glob import glob
from PIL import Image
```

```python
#read in paths of cropped photos
PRY_paths = glob('/Volumes/CW3E_data/CW3E_Streamflow_Archive/All_RR_Field_Cam_Photos/PRY_cropped/*') #get all the path names of the cropped PRY photos
print(PRY_paths)
print(len(PRY_paths))
```

```python
#create thumbnails and save them
save_path = '/Users/cloudy-guest/Desktop/RR/Photos/Thumb_PRY/'
MAX_SIZE = (75,90)
for i in PRY_paths[150:200]:
    str_split = i.split('/') #split path name so can get file name
    last_name = str_split[-1] #find file name
    img = Image.open(i)
    print(i)
    img.show()
    img.thumbnail(MAX_SIZE)
    img.save(save_path+last_name)
    
```
